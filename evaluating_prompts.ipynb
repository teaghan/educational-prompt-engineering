# Evaluating Prompts and Responses

## Having ChatGPT Rate its Responses

If we want to make the process of leveraging ChatGPT more efficient and dependable, we might be able to chain together multiple prompts and have the model assess its own work.

To do this, we will have to come up with some metrics for success, which will require some experimenting.

<div class="alert alert-info"><h4>Developing a set of Socials Studies discussion questions</h4><p>

**Prompt:**
    
You are a teacher with a passion for fostering open dialogue and critical thinking in your grade 10 social studies class. Your class just finished a unit on Globalization and its impact. Create three discussion questions that will encourage students to reflect on economic, cultural, and environmental aspects. For example: 

1. Reflecting on the interconnectedness of global economies, discuss how advancements in technology and communication might contribute to the interconnectedness of global economies. Use specific examples.

Format your three questions as a numbered list.
    
</p></div>

**ChatGPT:**

<font color='blue'>

1. Considering the cultural dimension of globalization, explore how the exchange of ideas, values, and traditions has been influenced by global interconnectedness. Provide examples of cultural interactions that have both positive and negative impacts, and discuss the implications for local cultures.

2. Investigate the environmental consequences of globalization by examining how the movement of goods, resources, and technologies on a global scale affects ecosystems and natural resources. Discuss specific instances where globalization has led to environmental challenges or innovations, and analyze the role of international cooperation in addressing these issues.

3. Delve into the economic disparities that can arise from globalization, particularly in relation to income inequality and access to resources. Explore how global economic interconnectedness may contribute to both economic growth and inequality within and between countries. Provide examples and discuss potential strategies for mitigating these disparities while still fostering economic growth on a global scale.

</font>

You can provide follow-up prompts to ChatGPT and the model will use the chat history to help determine its next response. 

We can leverage this feature by asking it to rate the three questions it provided. For the purposes of this investigation, I'm also interested in hearing its justification for these ratings.

</p></div>
